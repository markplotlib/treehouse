# Legends and Subplots

## Legends

````python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
````

````python
x = list(range(1, 101))
y = 50 * np.cos(25 * np.array(x)) - 100
z = np.tan((1/8) * np.array(x))

plt.plot(x, y, color='lime', linestyle='dotted', label='dotted')
plt.plot(x, z, color='olive', ls='dashed', label='dashed')
plt.ylabel('Key Findings')
plt.legend()
````

## subplots (to plot a grid)

````python
def demo_subplot(nrows, ncolumns):
    ### create first panel
    panel_num = 1
    panel_1 = plt.subplot(nrows, ncolumns, panel_num)
    plt.plot(x, y, color='magenta', ls=':')
    panel_1.set_ylim([-250, 50])

    ### create second panel
    panel_num = 2
    panel_2 = plt.subplot(nrows, ncolumns, panel_num)
    plt.plot(x, z, color='aqua', ls='-')
    panel_2.set_ylim([-250, 50])

    plt.show()
````

Notice that the `set_ylim()` method of `plt.subplot` sets each plot's y-limits, for easy comparison across plots

This will display: 2 rows, 1 column

````python
demo_subplot(2, 1)
````

This will display: 1 row, 2 columns

````python
demo_subplot(1, 2)
````
