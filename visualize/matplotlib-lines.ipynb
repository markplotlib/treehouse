```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

x = list(range(1,101))
y = np.array(x) ** 2
```

### LaTeX

$$ y = x^2 $$

Plot $y = x^2$

```python
plt.plot(x, y)
plt.ylabel('y = x^2')
plt.show()
```

## Adjust color and linestyle

### line color

[Matplotlib color options](https://matplotlib.org/api/colors_api.html)

The `color` parameter accepts:

* HTML color names
* hexcode values
* a tuple of RGB or RGBA values
* grayscale values: e.g., `'0.1'` (darker), `'0.9'` (lighter)
* and short code values.

### linestyles

From [matplotlib: lines](https://matplotlib.org/2.0.2/api/lines_api.html):

` ‘solid’ | ‘dashed’, ‘dashdot’, ‘dotted’ | (offset, on-off-dash-seq) |
'-' | '--' | '-.' | ':' | 'None' | ' ' | '' `


plt.plot(x, y, color='green', linestyle='-')

```python
plt.plot(x, np.sin(x), color='0.9', linestyle=':')
```

plt.plot(x, np.sqrt(x), color='0.1', linestyle='--')

plt.plot(x, np.log(x), color=(.1, .2, .5), linestyle='dashdot')

```python
plt.plot(x, -np.array(x)**2, color=('#00FF0F'), linestyle='dotted')
```

## multiple plots onto one set of axes

```python
plt.plot(x, 4000 * np.sin(50*np.array(x)), color=(.9, .2, .5), linestyle='dashed')
plt.plot(x, 6000 - np.array(x)**2, color=('#2B5B84'), linestyle='solid')
```
