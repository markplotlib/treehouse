# interactive mode

Interactive mode may also be turned on via 
`matplotlib.pyplot.ion()`
, and turned off via 
`matplotlib.pyplot.ioff()`.

### interactive backend example

````python
import matplotlib.pyplot as plt
plt.ion()
plt.plot([1.6, 2.7])
plt.title("interactive test")
plt.xlabel("index")
ax = plt.gca()
ax.plot([3.1, 2.2])
plt.draw()
````

````python
# now turn interactive mode off:
plt.ioff()
plt.plot([1.6, 2.7])
plt.show()
````

non-interactive mode delays all drawing until `show()` is called
